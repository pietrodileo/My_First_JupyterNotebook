### My code will display the first 20 terms of the Fibonacci sequence and plot them 

````python
# import module for plot
import matplotlib.pyplot as plt

nterms = 20
n1, n2 = 0, 1
idx = 0

# initialize x and y arrays
x = [*range(0,nterms,1)]
y = [1] * nterms

# generate Fibonacci sequence
while idx < nterms:
    y[idx] = n1
    nth = n1 + n2 
    # update values
    n1 = n2   
    n2 = nth
    idx += 1

# Print Fibonacci sequence
print(y)
print(x)

# plot Fibonacci on y axis and integers on x axis
plt.plot(x, y)

# name the axes
plt.xlabel('x')
plt.ylabel('Fibonacci Sequence')
  
# title of my plot
plt.title('My first plot!')
  
# show the plot
plt.show()
````
